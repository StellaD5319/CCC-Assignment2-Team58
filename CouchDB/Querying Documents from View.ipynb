## language of city

```python
def city_lang_comp(city:str):
    db_list = [f"{city}", f"{city}_tl7", "whole"]
    db_Au = client["whole"]
    db_current_city = client[f"{city}"]
    db_current_city_tl = client[f"{city}_tl7"]
    
    LANG_DIC = {}
    try:
        for item in db_Au.view('Doc/view_'+str(city)+'_lang', group=True):
            LANG_DIC[item.key] = item.value 

        for item in db_current_city.view('Doc/lang', group=True):
            LANG_DIC[item.key] = LANG_DIC[item.key] + item.value 

        for item in db_current_city_tl.view('Doc/lang', group=True):
            LANG_DIC[item.key] = LANG_DIC[item.key] + item.value 
    except:
        pass
    with open(str(city)+"_lang_comp.json","w+",encoding="UTF-8") as fp:
        json.dump(LANG_DIC, fp)
```

```python
Cities = ["Sydney", "Melbourne", "Brisbane", "Perth", "Adelaide"]
cities = ["sydney", "melbourne", "brisbane", "perth", "adelaide"]

for city in cities:
    city_lang_comp(city)
```
